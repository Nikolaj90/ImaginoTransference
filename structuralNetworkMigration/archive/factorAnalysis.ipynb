```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA, FactorAnalysis
import networkx as nx
import pandas as pd
```

```python
os.chdir("structuralNetworkMigration")
assert "data" in os.listdir()
```

```python
data = pd.read_csv("data/embeddings/from_graph_thresholds.csv")
```

```python
X = data.values
```

```python
ax = plt.axes()

im = ax.imshow(np.corrcoef(X))


plt.colorbar(im).ax.set_ylabel("$r$", rotation=0)
ax.set_title("correlation matrix")
plt.tight_layout()
```

```python
fa = FactorAnalysis(rotation="varimax")
fa.set_params(n_components=3)
fa.fit(X.T)
```

```python
components = fa.components_.T[150:,:]
```

```python
len(components)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
vmax = np.abs(components).max()
im = ax.imshow(components, cmap="RdBu_r")
plt.colorbar(im).ax.set_ylabel("$r$", rotation=0)
plt.show()
```
