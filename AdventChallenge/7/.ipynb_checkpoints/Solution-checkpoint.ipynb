```python
import regex as re
import networkx as nx
import numpy as np
import pandas as pd
```

```python
a = """$ cd /
$ ls
dir a
14848514 b.txt
8504156 c.dat
dir d
$ cd a
$ ls
dir e
29116 f
2557 g
62596 h.lst
$ cd e
$ ls
584 i
$ cd ..
$ cd ..
$ cd d
$ ls
4060174 j
8033020 d.log
5626152 d.ext
7214296 k
$ cd a
$ ls
dir e
29116 f
2557 g
62596 h.lst
$ cd e
$ ls
584 i
$ cd ..
$ cd ..
$ cd d
$ ls
4060174 j
8033020 d.log
5626152 d.ext
7214296 k"""

# a_list = a.split("\n")
```

```python
with open("input.txt") as f:
    a_list = f.read().split("\n")
```

```python
current_dir = "home/"
dir_dict = {}
embedded_dir = {}
log = []
network = []
nodes = {}
G = nx.DiGraph()


for cmd in a_list:
    # Match $ cd x
    if re.match("(?:\$ cd )([\./a-z]+)", cmd): 
        new_dir = re.findall("(?:\$ cd )([\./a-z]+)", cmd)
        
        # Hvis der gåes tilbage fjernes det yderste led fra stien    
        if new_dir[0] == "..":
            log.append(current_dir)
            current_dir = current_dir.rsplit(":", 1)[0]
            continue
            
        if re.match("[a-z/]+", new_dir[0]):
            
            # Dublet tjek, hvis det er en dublet node tilføjes den til dict og tælleren øges med en
            if new_dir[0] in G.nodes():
                new_node = new_dir[0]
                i = 0
                while new_node in G.nodes():
                    i += 1
                    new_node = new_dir[0] + f"_{i}"

            if new_dir[0] not in G.nodes():
                new_node = new_dir[0]
                
            current_dir = current_dir + f":{new_node}"

            # Nu tilføjes forbindelsen, som edge i mit netværk
            parent_node = current_dir.rsplit(":",2)[1]

            edge = (parent_node,new_node)
            print(edge)
            G.add_edge(parent_node,new_node)

            
    # Matcher på 10000 xxx        
    if re.match("(\d+)\s\w+", cmd):
        # det yderste led af stien separeres 
        dir_ = current_dir.rsplit(":", 1)[1]
        # fil størrelsen isoleres
        file_size = [int(re.findall("(\d+)\s\w+", cmd)[0])]
        
        if dir_ in dir_dict.keys():
            dir_dict[dir_].extend(file_size)
        if dir_ not in dir_dict.keys():
            dir_dict[dir_] = file_size
            
```

```python
labels = {n: n for n in G.nodes()}
nx.draw(G, with_labels = True, labels = labels)
```

```python
for node in G.nodes:
    if node in dir_dict.keys():
        G.nodes[node]["weight"] = np.array(dir_dict[node]).sum()
    else:
        G.nodes[node]["weight"] = 0
```

```python
list_df = list(G.nodes)
weights_df = [G.nodes[node]["weight"] for node in list_df]

df = pd.DataFrame(zip(list_df,weights_df))
df.head()

# alle får 0 som startværdi, hvis de overskrider de 100000, hvis de er gennemgået og under 100000 får de True som værdi i 
# en separat kolonne
```

```python
weights_total = []

for node in df[0]:
    weight = np.array([G.nodes[child_node]["weight"] for child_node in [*nx.bfs_tree(G, node)]]).sum()
    weights_total.append(weight)
```

```python
df["weights_total"] = weights_total
df[df["weights_total"]<=100000]["weights_total"].sum()
```

```python
# Part 2
DISK_SPACE = 70000000
UPDATE_SPACE = 30000000

free_space = DISK_SPACE - max(df["weights_total"])

df[df["weights_total"] >= (UPDATE_SPACE - free_space)].sort_values("weights_total")
```
